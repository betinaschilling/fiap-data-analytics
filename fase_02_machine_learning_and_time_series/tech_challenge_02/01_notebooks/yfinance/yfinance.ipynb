```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Lista os índices e Simbolos no Yahoo Finance 

indices = {
    "Ibovespa": "^BVSP",
    "Dow Jones Industrial Average (DJIA)": "^DJI",
    "S&P 500": "^GSPC",
    "NASDAQ Composite": "^IXIC",
    "FTSE 100": "^FTSE",
    "DAX": "^GDAXI",
    "CAC 40": "^FCHI",
    "Nikkei 225": "^N225",
    "Hang Seng": "^HSI",
    "Shanghai Composite": "000001.SS",
    "BSE Sensex": "^BSESN"
}
```

```python
# Dicionário para armazenar os DataFrames
data_frames = {}
```

```python
# Busca os dados históricos dos índices e armazena em DataFrames
for index_name, ticker in indices.items():
    index_data = yf.Ticker(ticker)
    hist = index_data.history( start = '1994-01-01', end = '2024-07-01')  # Buscar máximo de dados
    data_frames[index_name] = hist["Close"] # Armazenar apenas a coluna de fechamento
    #data_frames[index_name] = hist  # Armazenar o DataFrame no dicionário
    print(f"Data for {index_name} ({ticker}) fetched successfully!")
```

```python
# Para acessar um DataFrame específico
ibovespa_df = data_frames["Ibovespa"]
print(ibovespa_df.head())
```

```python
# Para Salvar os DataFrames em arquivos CSV (opcional)
for index_name, df in data_frames.items():
    df.to_csv(f"{index_name}.csv")
```

```python
# Criar um DataFrame combinado para comparação
combined_df = pd.DataFrame(data_frames)
```

```python

# Plotar os gráficos de comparação
plt.figure(figsize=(14, 7))
for index_name in combined_df.columns:
    plt.plot(combined_df.index, combined_df[index_name], label=index_name)

plt.title("Comparação dos Principais Índices de Ações")
plt.xlabel("Data")
plt.ylabel("Fechamento")
plt.legend()
plt.grid(True)
plt.show()
```
